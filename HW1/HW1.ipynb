```python
import random
```

```python
def rules(n):
    while n != '' and n != '1':
        n = input('Что-то пошло не так. Может быть, вы хотели ввести 1 и вывести правила?\n Чтобы вывести правила игры, введите 1. Если вы знаете правила, нажмите Enter.')
    if n == '':
        return
    elif n == '1':
        print('Загадывается слово — программа отмечает места букв чертами и рисует виселицу. \n\
        Согласно традиции русских лингвистических игр слово должно быть именем существительным нарицательным \n\
        в именительном падеже единственного числа, либо множественного числа при отсутствии у слова формы единственного числа. \n\
        Игрок предлагает букву, которая может входить в это слово. \n\
        Если такая буква есть в слове, то программа заменяет соответствующие этой букве черточки столько раз, \n\
        сколько буква встречается в слове. Если такой буквы нет, то к виселице добавляется круг, изображающий голову. \n\
        Игрок продолжает отгадывать буквы до тех пор, пока не отгадает всё слово. \n\
        За каждый неправильный ответ программа добавляет одну часть туловища к виселице (всего 15 частей). \n\
        Если туловище в виселице нарисовано полностью, то игрок проигрывает, считается повешенным. \n\
        Если игроку удаётся угадать слово, он выигрывает. Приятной игры!')
```

```python
# Функция запрашивает номер темы у пользователя и выводит его или индикатор некорректности ввода.

def сhoose_theme():
    n = input('Выберите одну из трёх предложенных тем игры: Персонажи Туве Янссон (введите 1), Названия созвездий (введите 2) или Города (тауны) графств Корк и Голуэй в Ирландии (введите 3):\n')
    if n == '1' or n == '2' or n == '3':
        return int(n)
    else:
        return 0
```

```python
# Функция запрашивает номер темы и выводит список из возможных слов для загадывания.

def form_list(n):
    filename = "theme" + str(n) + ".txt"
    with open(filename, "r", encoding="utf-8") as f:
        l = f.read().splitlines()
        return l
```

```python
# Функция рисует виселицу исходя из кол-ва неуспешных попыток.

def draw_scaffold(i):
    scaffold = ['____','\n|  |','\n|  ','\n|  ','\n| ','\n|  ','\n|  ']
    manikin = [['(', 2], ['x',2], ['_',2], ['x',2], [')',2],
              ['|   ',3], ['|',3], ['/',4], ['|',4], ['_',4], ['_',4], ['_',4], ['|',4], ['\\',4],
              ['|   ',5], ['|',5]]
    drawing = scaffold
    for j in range(i):
        if i == 0:
            break
        s = manikin[j]
        drawing[s[1]] += s[0]
    print(''.join(drawing))
```

```python
# Функция выводит "У вас есть N попыток, чтобы отгадать слово из X букв".

def check_attempts(N, X, l):
    if N == 1:
        print("У вас есть 1 попытка, чтобы отгадать слово из ", X, " букв.")
    elif N < 5:
        print("У вас есть ", N, " попытки, чтобы отгадать слово из ", X, " букв.")
    else:
        print("У вас есть ", N, " попыток, чтобы отгадать слово из ", X, " букв.")
    draw_scaffold(16-(N+1))
    print(''.join(l))
```

```python
# Функция запрашивает уже введённые ранее буквы, а также новую букву и проверяет её на адекватность и новизну.

def check_letter(attempts):
    while True:
        L = input('Введите букву кириллицей.')
        L = L.lower()
        if L in list('йцукенгшщзхъфывапролджэячсмитьбюё'):
            if L in attempts:
                print('Вы уже проверяли эту букву. Попробуйте другие.')
            else:
                attempts += L
                return attempts, L
        else:
            print('Введённый символ не похож на кириллическую букву. Попробуйте ещё.')
```

```python
# Функция запрашивает введённую букву, загаданное слово и список угаданных букв 
# и возвращает информацию об (не)успешности хода и обновлённый список.

def check_word(L, word, l):
    if L in word.lower():
        for i, element in enumerate(word.lower()):
            if element == L:
                l[i] = L
        return True, l
    else:
        return False, l
```

```python
def more_letters(l):
    if '_' in l:
        return True
```

```python
# Функция спрашивает у пользователя, повторить ли игру.

def again():
    n = input("Сыграем ещё раз?\n Введите любой набор символов, чтобы продолжить. Для выхода нажмите Enter.")
    if n == '':
        return True
    else:
        return False
```

```python
# Главная функция, в которой происходят вызовы вспомогательных функций.

def main():
    rules(input('Добро пожаловать в игру "Виселица"!\n Чтобы вывести правила игры, введите 1. Если вы знаете правила, нажмите Enter.'))
    while True:
        theme_number = сhoose_theme()
        while theme_number == 0:
            print('Вы ввели некорректный номер темы. Попробуйте ещё раз.')
            theme_number = сhoose_theme()
        word = random.choice(form_list(theme_number))
        X = len(word)
        N = 16
        l = ['_' for i in range(X)]
        attempts = ''
        check_attempts(N, X, l)
        while N > 0:
            if attempts != '':
                check_attempts(N, X, l)
            attempts, L = check_letter(attempts)
            N -= 1
            is_in_word, l = check_word(L, word, l)
            if is_in_word:
                print('Повезло! В этом слове действительно есть такая буква!')
                if not more_letters(l):
                    print('Ура! Вы угадали все буквы! Загаданное слово - ', word, '\nЭто было не так-то просто, вы восхитительны!')
                    N = 0
            else:
                print('Неудача... Такой буквы в слове нет.')
                if N == 0:
                    print('У вас закончились попытки...') 
                    draw_scaffold(16)
                    print('Но в следующий раз вы обязательно угадаете! Ну а загаданное слово было ', word)
        if again():
            break
```

```python
if __name__=="__main__":
    main()
```
