```python
from string import punctuation
```

```python
# Функция запрашивает имя файла, открывает его
# и создаёт множество из слов в файле.
def read_file(filename):
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read()
        words = set([i.strip(punctuation) for i in text.split()])
    return words
```

```python
def main():
    words1 = read_file('Blok.txt')
    words2 = read_file('Gumilev.txt')
    common = words1 & words2
    unique = words1 ^ words2
    print(common, "\n\n", unique)
```

```python
if __name__ == "__main__":
    main()
```
